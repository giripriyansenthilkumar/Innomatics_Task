1. User Login Check

```python
username = "admin"
password = "1234"
user=input()
passw=input()
if user==username and passw==password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```


2. Pass / Fail Analyzer


```python
marks=list(map(int,input().split(",")))
pass_s=0
fail_s=0
for i in range(len(marks)):
    if marks[i]>=50:
        pass_s+=1
    else:
        fail_s+=1
print(f"Passed Students : {pass_s}")
print(f"Failed Students : {fail_s}")
```

3. Simple Data Cleaner

```python
names=list(map(str,input().split(",")))
for i in range(len(names)):
    s=names[i].replace(" ","")
    s=names[i].lower()
    names[i]=s
print(names)
```

4. Message Length Analyzer

```python
messages=list(map(str,input().split(",")))
flag=0
length=[]
for i in range(len(messages)):
    l=len(messages[i])
    if l>10:
        flag+=1
    length.append(l)
print("Length of each messages :",length)
print("Number of flagged messages :",flag)
```

5. Error Message Detector

```python
logs=list(map(str,input().split(",")))
c=0
for i in range(len(logs)):
    if logs[i].upper()=="ERROR":
        c+=1
    # print(logs[i])
print(c)
```
