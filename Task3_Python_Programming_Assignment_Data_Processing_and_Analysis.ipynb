Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees={"Ravi":92,"Anita":88,"Kiran":92,"Suresh":85}
scores=list(employees.values())
highest=scores[0]
for s in scores:
    if s>highest:
        highest=s
eligible=[]
for emp in employees:
    if employees[emp]==highest:
        eligible.append(emp)
print("Top Performers Eligible for Bonus:",end=" ")
for i in range(len(eligible)):
    if i!=len(eligible)-1:
        print(eligible[i]+",",end=" ")
    else:
        print(eligible[i],end=" ")
print("(Score:",highest,")")
```

Problem Statement 2: Search Query Keyword Analysis

```python
text="Buy mobile phone buy phone online"
clean_text=""
for ch in text:
    if ch.isalpha() or ch==" ":
        clean_text+=ch.lower()
words=clean_text.split()
count_map={}
for w in words:
    if w in count_map:
        count_map[w]+=1
    else:
        count_map[w]=1
duplicate_words={}
for key in count_map:
    if count_map[key]>1:
        duplicate_words[key]=count_map[key]
print(duplicate_words)
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings=[3,4,7,8,10,12,5]
valid_data=[]
index=0
for value in sensor_readings:
    if value%2==0:
        valid_data.append((index,value))
    index+=1
print("Valid Sensor Readings (Hour, Value):")
print(valid_data)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails=["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domains=[]
for mail in emails:
    parts=mail.split("@")
    domains.append(parts[1])
domain_freq={}
for d in domains:
    if d in domain_freq:
        domain_freq[d]+=1
    else:
        domain_freq[d]=1
total=0
for _ in emails:
    total+=1
for dom in domain_freq:
    percent=(domain_freq[dom]*100)/total
    print(dom+":",str(int(percent))+"%")
```

Problem Statement 5: Sales Spike Detection

```python
sales=[1200,1500,900,2200,1400,3000]
total=0
for s in sales:
    total+=s
average=total/len(sales)
limit=average+(average*30/100)
day=1
for value in sales:
    if value>limit:
        print("Day",day,":",value)
    day+=1
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids=["user1","user2","user1","user3","user1","user3"]
track={}
for i in user_ids:
    if i in track:
        track[i]+=1
    else:
        track[i]=1
for key in track:
    if track[key]>1:
        print(key,"→",track[key],"times")
```
