Problem 1: Mobile Recharge Validation System

```python
def validate():
    plans=[199,299,399,599]
    while True:
        amount=int(input())
        if amount<50:
            print("Recharge amount should be greater than or equal to 50")
            continue
        if amount in plans:
            print(f"Recharge Successful for Rs.{amount}")
            break
        else:
            print("Invalid Plan")
            print("Retry Again")
validate()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_data):
    for product in stock_data:
        if stock_data[product]<15:
            print(product,":Reorder Alert")
        else:
            print(product,":Stock OK")

products = {"Rice":10,"Wheat":25,"Sugar":12,"Oil":30}
check_inventory(products)
```

 Problem 3: Student Result Processing System

```python
def result():
    marks=list(map(int,input().split(",")))
    print(f"Marks:{marks}")
    print((sum(marks)//len(marks)))
    for i in range(len(marks)):
        if marks[i]>=50:
            print("Pass")
        else:
            print("Fail")

result()
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def cab_fare(distance,peak):
    base=50
    per_km=12
    fare=base+(distance*per_km)
    if peak:
        fare+=(fare*0.25)
    return fare

while True:
    distance=float(input())
    peak=bool(input())
    total=cab_fare(distance,peak)
    print("Total Fare:",total)
    exit=input("Need to exit?").lower()
    if exit=="yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_check(attendence):
    present=0
    for day in attendance:
        if day=="P":
            present+=1
    total_days=len(attendance)
    percentage=(present/total_days)*100
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance =list(map(str,input().split()))
status=attendance_check(attendance)
print("Attendance Status:",status)
```

Problem 6: Password Strength Checker

```python
def password(s):
    has_digit=False
    has_special=False
    special_chars="@#$"
    if len(s)<8:
        return "Weak Password"
    for ch in s:
        if ch.isdigit():
            has_digit=True
        if ch in special_chars:
            has_special=True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

s=input()
print(password(s))
```
